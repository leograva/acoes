```python
from selenium import webdriver
import time
import os
import xlsxwriter
from datetime import date

#link do statusinvest - iremos só adicionar no final a sigla da ação
link = "https://statusinvest.com.br/acoes/"
link2 = "https://statusinvest.com.br/fundos-imobiliarios/"

#ler arquivo contendo nome das ações e transformando num array
arquivo = open('acoes2.txt', 'r')
lista = arquivo.readlines() # readlinesssssss
arquivo.close()

#ler arquivo contendo nome dos fii's e transformando num array
arquivo = open('fiis2.txt', 'r')
lista2 = arquivo.readlines() # readlinesssssss
arquivo.close()

#setando oconfigurações do webdriver 
options = webdriver.ChromeOptions()
options.add_argument('lang=pt-br')
options.add_argument("--headless")
driver = webdriver.Chrome(executable_path=r'./chromedriver', options=options)
driver.set_window_position(-10000,0)

#importando data
data_atual = date.today()
data_em_texto = data_atual.strftime('%d/%m/%Y')

#concatenando data para criar o nome do arquivo a ser gerado
dataparauso = "valores_"+data_em_texto.replace('/','_')+".xlsx"


# criar arquivo excel
workbook = xlsxwriter.Workbook(dataparauso)
worksheet = workbook.add_worksheet("acoes")
worksheet2 = workbook.add_worksheet("fiis")

#montando cabeçalho ações
worksheet.write("A1","Ação");
worksheet.write("B1","Valor");

#montando cabeçalho fiis
worksheet2.write("A1","FII");
worksheet2.write("B1","Valor");


def getPrecoAcoes():
    count = 2
    print("Preço atual das ações:")
    for i in range(len(lista)):
            driver.get(link+lista[i])
            chat_box = driver.find_element_by_xpath("//*[@id='main-2']/div[2]/div/div[1]/div/div[1]/div/div[1]/strong").text
            
            print(lista[i].rstrip('\n'),"- R$",chat_box)
            valorcelula_a = 'A'+str(count);
            
            lisval = str(lista[i])
            worksheet.write(valorcelula_a,lisval);
            
            valorcelula_b = 'B'+str(count);
            valor = float(chat_box.replace(',','.'))
            
            worksheet.write(valorcelula_b,valor);
            count=count+1

def getPrecoFiis():
    count2 = 2
    print("Preço atual dos FII'S")
    for i in range(len(lista2)):
            driver.get(link2+lista2[i])
            chat_box2 = driver.find_element_by_xpath("//*[@id='main-2']/div[2]/div[1]/div[1]/div/div[1]/strong").text
            
            print(lista2[i].rstrip('\n'),"- R$",chat_box2)
            valorcelula_a = 'A'+str(count2);
            
            lisva2 = str(lista2[i])
            worksheet2.write(valorcelula_a,str(lista2[i]));
            
            valorcelula_b = 'B'+str(count2);
            valor2 = float(chat_box2.replace(',','.'))
            
            worksheet2.write(valorcelula_b,valor2);
            count2=count2+1
        
def finalizarArquivoXlsx():
    workbook.close()
    print("\nArquivo xlsx salvo!!!")

#executando métodos    
getPrecoAcoes();
print("\n")
getPrecoFiis();
finalizarArquivoXlsx();
```
